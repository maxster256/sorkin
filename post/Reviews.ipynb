```python
import pandas as pd
```

```python
df = pd.read_pickle("datasets/rated_reviews.gzip")
```

```python
df = df.drop(["_id"], axis=1)
df.head(5)
```

```python
df.describe()
```

```python
df = df.astype({'rating': 'int32'})
df.dtypes
```

```python
def add_sentiment():
    """
    Adds a sentiment class to the dataset based on the rating given by the reviewer.
    """
    def determine_sentiment(row):
        if row['rating'] in [1, 2, 3, 4]:
            return "negative"
        elif row['rating'] in [5, 6, 7]:
            return "neutral"
        elif row['rating'] in [8, 9, 10]:
            return "positive"
            
    df['sentiment'] = df.apply(lambda row: determine_sentiment(row), axis=1)
```

```python
add_sentiment()
```
